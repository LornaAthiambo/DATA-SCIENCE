```python
#(Q2) Implementation of coordinate subtraction

class Coordinate:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def subtract(self, other):
        return Coordinate(self.x - other.x, self.y - other.y)
    
    # Create two instances of the Coordinate class
coord1 = Coordinate(7, 9)
coord2 = Coordinate(4, 6)
result = coord1.subtract(coord2)

print(result.x, result.y)


```

```python
# (Q1) solving correlation coefficient of a known factor
# Assuming you have the factors 
x = np.array([10, 30, 50, 70, 100])
y = np.array([20, 40, 60, 80, 90])

# Calculate the product correlation coefficient
n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)
x_std = np.std(x)
y_std = np.std(y)

numerator = np.sum(x * y) - n * x_mean * y_mean
denominator = np.sqrt((np.sum(np.power(x, 2)) - n * np.power(x_mean, 2)) * (np.sum(np.power(y, 2)) - n * np.power(y_mean, 2)))

product_correlation = numerator / denominator

print(product_correlation)
```

```python
#(Q3)Writing email code

import re

def split_email_regex(email):
    match = re.search(r"([^@]+)@(.+)", email)
    if match:
        username, domain = match.groups()
        return username, domain
    else:
        return None, None

user_email = input("Enter your email address: ")
username, domain = split_email_regex(user_email)
if username and domain:
    print(f"Username: {username}")
    print(f"Domain: {domain}")
else:
    print("Invalid email address format.")
```

```python
#(Q4) Quadratic equation

import math

def calculate_roots(a, b, c):
    discriminant = b**2 - 4*a*c

    if discriminant < 0:
        print("The equation has no real roots.")
    elif discriminant == 0:
        root = -b / (2*a)
        print(f"The equation has a single root: {root}")
    else:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        print(f"The equation has two roots: {root1} and {root2}")

# Example usage
a = float(input("Enter the value of a: "))
b = float(input("Enter the value of b: "))
c = float(input("Enter the value of c: "))

calculate_roots(a, b, c)
```
